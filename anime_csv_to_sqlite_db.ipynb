# Anime CSV to SQLite DB
This notebook converts the anime.csv file into a SQLite database.

## Note:
This script was modified from the original `main.ipynb` notebook from the [Anime Database - Combining Crunchyroll and MyAnimeList
 repo](https://github.com/cckuqui/anime-db) to work with the [anime.csv](https://www.kaggle.com/datasets/shafiwalsher/myanimelist-dataset-2023-top-15000) file.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('animes.csv', encoding='utf-8')
df
```

```python
df.info()
```

```python
# drop trailer, trailer_thumbnail, premiered, synopsis_y
df.drop(['trailer', 'trailer_thumbnail', 'premiered', 'synopsis_y'], axis=1, inplace=True)
# drop null aired
df.dropna(subset=['aired'], inplace=True)
# drop null genres
df.dropna(subset=['genres'], inplace=True)
```

```python
df.info()
```

```python
# print all nan data in genres
# df[df['genres'].isnull()]
```

```python
title = []
episodes = []
score = []
link = []
genre = []
synopsis = []
duration = []
aired = []
members = []
rating = []
status = []
studio = []
titletype = []
```

```python
for index, data in df.iterrows():
    data_genre = data['genres'] 
    data_genre = data_genre.split(',')
    
    for dg in data_genre:
        genre.append(dg)
        title.append(data['title'])
        episodes.append(data['episodes'])
        score.append(data['score'])
        link.append(data['poster'])
        synopsis.append(data['synopsis_x'])
        duration.append(data['duration'])
        aired.append(data['aired'])
        members.append(data['members'])
        rating.append(data['rating'])
        status.append(data['status'])
        studio.append(data['studios'])
        titletype.append(data['titletype'])
```

```python
dict = {
        'title': title,
        'link': link,
        'episodes': episodes,
        'score': score,
        'genre': genre,
        'synopsis': synopsis,
        'duration': duration,
        'aired': aired,
        'members': members,
        'rating': rating,
        'status': status,
        'studio': studio,
        'title_type': titletype
        } 
myal = pd.DataFrame(dict)
```

```python
myal.head()
```

```python
# Trim genre
myal['genre'] = myal['genre'].str.strip()
# Replace unknown episodes with null
myal['episodes'] = myal['episodes'].replace('Unknown', np.nan)
# Replace unknown duration with null
myal['duration'] = myal['duration'].replace('Unknown', np.nan)
# Replace 'None found,' studio with null
myal['studio'] = myal['studio'].replace('None found,', np.nan)
```

```python
myal.head()
```

```python
anime = myal[['title','episodes','link','score','synopsis', 'duration', 'aired', 'members', 'rating', 'status', 'studio', 'title_type']]
anime.drop_duplicates('title', inplace = True)
anime.reset_index(drop=True, inplace=True)
anime['id'] = anime.index + 1
anime
```

```python
genre = myal[['genre']]
genre = genre.drop_duplicates('genre')
genre = genre.reset_index(drop=True)
genre['id'] = genre.index + 1
genre
```

```python
import sqlite3
```

```python
create_table_query="""
BEGIN TRANSACTION;
CREATE TABLE IF NOT EXISTS "user" (
	"id" INTEGER PRIMARY KEY,
	"username" TEXT NOT NULL UNIQUE,
	"password" TEXT NOT NULL,
	"profile_picture_link" TEXT,
	"bio_description" TEXT,
	"created_at" DATETIME DEFAULT CURRENT_TIMESTAMP,
	"updated_at" DATETIME DEFAULT CURRENT_TIMESTAMP,
	"account_type" TEXT NOT NULL DEFAULT 'USER'
);
CREATE TABLE IF NOT EXISTS "list" (
	"id" INTEGER PRIMARY KEY,
	"user_id" INTEGER NOT NULL,
	"name" TEXT NOT NULL,
	"description" TEXT,
	"created_at" TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
	"updated_at" TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    UNIQUE ("user_id", "name"),
	FOREIGN KEY("user_id") REFERENCES "user"("id") ON DELETE CASCADE
);
CREATE TABLE IF NOT EXISTS "list_anime" (
	"list_id" INTEGER NOT NULL,
	"anime_id" INTEGER NOT NULL,
	"added_at" TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
	PRIMARY KEY("list_id","anime_id"),
	FOREIGN KEY("anime_id") REFERENCES "anime"("id") ON DELETE CASCADE,
	FOREIGN KEY("list_id") REFERENCES "list"("id") ON DELETE CASCADE
);
CREATE TABLE IF NOT EXISTS "user_anime_status" (
	"status" CHAR(1) PRIMARY KEY,
	"seq" INTEGER NOT NULL
);
CREATE TABLE IF NOT EXISTS "anime" (
	"id" INTEGER PRIMARY KEY,
	"title" TEXT,
	"episodes" TEXT,
	"link" TEXT,
	"score" REAL,
	"synopsis" TEXT,
    "duration" TEXT,
    "aired" TEXT,
    "members" INTEGER,
    "rating" TEXT,
    "status" TEXT,
    "studio" TEXT,
    "title_type" TEXT,
    "created_at" TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
CREATE TABLE IF NOT EXISTS "genre" (
	"id" INTEGER PRIMARY KEY,
	"genre" TEXT UNIQUE
);
CREATE TABLE IF NOT EXISTS "anime_genre" (
	"id" INTEGER PRIMARY KEY,
	"anime_id" INTEGER,
	"genre_id" INTEGER,
	FOREIGN KEY("anime_id") REFERENCES "anime"("id") ON DELETE CASCADE,
	FOREIGN KEY("genre_id") REFERENCES "genres"("id") ON DELETE CASCADE
);
CREATE TABLE IF NOT EXISTS "user_anime" (
	"user_id" INTEGER NOT NULL,
	"anime_id" INTEGER NOT NULL,
	"review" TEXT,
	"score" INTEGER,
	"watched_episodes" INTEGER,
	"status" CHAR(1) DEFAULT ('W'),
	PRIMARY KEY("user_id","anime_id"),
	FOREIGN KEY("anime_id") REFERENCES "anime"("id") ON DELETE CASCADE,
	FOREIGN KEY("user_id") REFERENCES "user"("id") ON DELETE CASCADE,
	FOREIGN KEY("status") REFERENCES "user_anime_status"("status")
);

INSERT OR IGNORE INTO "user_anime_status" VALUES ('W',1);
INSERT OR IGNORE INTO "user_anime_status" VALUES ('C',2);
INSERT OR IGNORE INTO "user_anime_status" VALUES ('O',3);
INSERT OR IGNORE INTO "user_anime_status" VALUES ('D',4);
INSERT OR IGNORE INTO "user_anime_status" VALUES ('P',5);

COMMIT;
"""
```

```python
conn = sqlite3.connect('anime.db')
c = conn.cursor()
```

```python
# conn.close()
```

```python
c.executescript(create_table_query)
conn.commit()
```

```python
anime.to_sql('anime', con=conn, if_exists='append', index=False)
genre.to_sql('genre', con=conn, if_exists='append', index=False)
```

```python
merge_df_ag = myal[['title', 'genre']]
merge_df_ag.drop_duplicates(inplace=True)
merge_df_ag
```

```python
# Get all rows with title starting with Bocchi
# merge_df_ag[merge_df_ag['title'].str.startswith('Bocchi')]
```

```python
# get values from anime table and keep id as column

table_anime = pd.read_sql_query('select * from anime', con=conn).reset_index()
table_anime
```

```python
table_genre = pd.read_sql_query('select * from genre', conn).reset_index()
table_genre.head()
```

```python
merge_df_ag = merge_df_ag.merge(table_anime, on='title')
merge_df_ag.rename(columns={'index':'anime_id'}, inplace=True)
merge_df_ag = merge_df_ag.merge(table_genre, left_on='genre', right_on='genre')
merge_df_ag.rename(columns={'index':'genre_id'}, inplace=True)
merge_df_ag
```

```python
# Get all rows with title starting with Bocchi
merge_df_ag[merge_df_ag['title'].str.startswith('Bocchi')]
```

```python
genre_anime = merge_df_ag[['anime_id','genre_id']]
genre_anime.drop_duplicates()
# Increment id_anime and id_genre by 1
genre_anime['anime_id'] += 1
genre_anime['genre_id'] += 1
genre_anime['id'] = genre_anime.index + 1
genre_anime.head()
```

```python
genre_anime.to_sql(name='anime_genre', con=conn, if_exists='append', index=False)
```
